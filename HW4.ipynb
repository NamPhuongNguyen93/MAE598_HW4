```python
import numpy as np

def f(x1, x2, x3):
    return x1 ** 2 + x2 ** 2 + x3 ** 2

def e_constraint(x1, x2, x3):
    h1 = x1 ** 2 / 4 + x2 ** 2 / 5 + x3 ** 2 / 25 - 1
    h2 = x1 + x2 - x3
    return h1, h2

def df_dx(x1, x2, x3):
    df_ds = np.vstack((2 * x2, 2 * x3))
    df_dd = np.array([2 * x1])
    return df_ds, df_dd

def dh_dx(x1, x2, x3):
    dh_ds = np.vstack((np.hstack(((2/5) * x2, (2/25) * x3)), [1, -1]))
    dh_ds_inv = np.linalg.inv(dh_ds)
    dh_dd = np.vstack(((1 / 2) * x1, 1.))
    return dh_ds_inverse, dh_dd, dh_ds

def df_dd(x1, x2, x3):
    df_ds, pf_pd = df_dx(x1, x2, x3)
    dh_ds_inverse, dh_dd, _ = dh_dx(x1, x2, x3)
    return pf_pd - np.matmul(df_ds.T, np.matmul(dh_ds_inverse, dh_dd))

def solve(x1, x2, x3):
    error = 1e-3
    h1, h2 = e_constraint(x1, x2, x3)
    h = np.vstack((h1, h2))
    hnorm = np.linalg.norm(h)
    while hnorm >= error:
        dh_inverse, _, dh = dh_dx(x1, x2, x3)
        ds = np.matmul(dh_inverse, h)
        x2 = x2 - ds[0]
        x3 = x3 - ds[1]
        h1, h2 = e_constraint(x1, x2, x3)
        h = np.vstack((h1, h2))
        hnorm = np.linalg.norm(h)
    return x1, x2, x3, hnorm

def linesearch(x1, x2, x3):
    a = 1
    df = df_dd(x1, x2, x3)
    phi = lambda a, x1, x2, x3, df: f(x1, x2, x3) - a * 0.3 * np.matmul(df, df.T) 

    def f_a(x1, x2, x3, a):
        df = df_dd(x1, x2, x3)
        dh_ds_inverse, dh_dd, _ = dh_dx(x1, x2, x3)

        x1 = x1 - a * df.flatten()
        ds = np.matmul(np.matmul(dh_ds_inverse, dh_dd), df.T).flatten()
        x2 = x2 + a * ds[0]
        x3 = x3 + a * ds[1]
        return f(x1, x2, x3)

    while phi(a, x1, x2, x3, df) < f_a(x1, x2, x3, a):
        a = 0.5 * a
        df = df_dd(x1, x2, x3)
    return a


def optimization(x1,x2,x3):
    
    epsilon =0.001 
    df_dd_norm = np.linalg.norm(df_dd(x1, x2, x3))

    while df_dd_norm >= epsilon: 
    
        a = linesearch(x1,x2, x3)
        x1 = x1 - a * df_dd(x1, x2, x3).flatten()
        dh_ds_inverse, dh_dd, _ = dh_dx(x1, x2, x3)
        ds = a * np.matmul(np.matmul(dh_ds_inverse, dh_dd), df_dd(x1, x2, x3).T).flatten()
        x2 = x2 + ds[0]
        x3 = x3 + ds[1]
        x1, x2, x3, _ = solve(x1,x2, x3)
        df_dd_norm = np.linalg.norm(df_dd(x1, x2, x3))
        
    return x1,x2,x3
        

x1 = 1.0
x2 = 1.0
x3 = 2.0

x1 = solve(x1, x2, x3)[0]
x2 = solve(x1, x2, x3)[1]
x3 = solve(x1, x2, x3)[2]

x1,x2,x3 = optimization(x1,x2,x3)


print('X=({},{},{})'.format(x1, x2, x3))
print('f_min is',f(x1,x2,x3))
```
